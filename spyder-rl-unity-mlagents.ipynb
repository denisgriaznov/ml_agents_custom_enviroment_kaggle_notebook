## Check mlagents ant Gym enviroment

```python
from mlagents_envs.environment import UnityEnvironment
from mlagents_envs.envs.unity_gym_env import UnityToGymWrapper
```

```python
import numpy as np
```

```python
env_path = "/kaggle/working/spyder_walker_linux/AI Spyder.x86_64"
!chmod -R 755 "/kaggle/working/spyder_walker_linux/AI Spyder.x86_64"
```

```python
def get_env():
    unity_env = UnityEnvironment(env_path, no_graphics=True)
    env = UnityToGymWrapper(unity_env)
    return env
```

```python
env = get_env()
print("Action space: ", env.action_space)
print("State space: ", env.observation_space)
env.reset()
for i in range(5):
    state, reward, done, _ = env.step(np.ones(12))
    print(reward, " ", done)
env.close()
```

## Learning with Stable Baselines 3 Soft-Actor Critic

```python
from stable_baselines3 import SAC
```

```python
env = get_env()
model = SAC("MlpPolicy", env, verbose=1, learning_rate=0.001, tensorboard_log="./sac_spyder_tensorboard/")
model.learn(total_timesteps=1000, log_interval=4)
model.save("sac_model.zip")
env.close()
```
